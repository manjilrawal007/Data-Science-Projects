## 1. Setup

```python

import pandas as pd
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", 100)
pd.set_option("display.width", 120)
```

## 2. Load Dataset

```python

# Load dataset
df = pd.read_csv("Project_data.csv")
df.head()
```

## 3. Visualization 1

```python

# Visualization 1 — Histogram of Daily Social Media Usage

plt.figure(figsize=(8,5))
plt.hist(df["Avg_Daily_Usage_Hours"], bins=15)
plt.xlabel("Average Daily Usage (Hours)")
plt.ylabel("Count")
plt.title("Distribution of Students' Daily Social Media Usage")
plt.savefig('Manjil_platforms_histogram.jpg')
plt.show()
```


### Interpretation of Visualization 1
The histogram shows that most students use social media between **4 and 7 hours per day**.
This indicates high daily engagement, suggesting potential academic or behavioral impacts.


## 4. Visualization 2

```python

# Visualization 2 — Most Used Social Media Platforms

platform_counts = df["Most_Used_Platform"].value_counts()

plt.figure(figsize=(8,5))
plt.bar(platform_counts.index, platform_counts.values)
plt.xlabel("Platform")
plt.ylabel("Number of Students")
plt.title("Most Commonly Used Social Media Platforms")
plt.xticks(rotation=30)
plt.savefig('Manjil_platforms_barchart.jpg')
plt.show()
```


### Interpretation of Visualization 2
Instagram and TikTok appear to be the most frequently used platforms.
These platforms may contribute significantly to total daily usage hours and potential addiction scores.



## Final Summary
These visualizations provide two key insights:
1. Students spend a considerable amount of time daily on social media, mostly 4–7 hours.
2. A few platforms dominate student usage, highlighting potential focus areas for further analysis.
